```python
import nltk
import sys 
import codecs
from nltk.corpus import stopwords
```

```python
from wordcloud import WordCloud
nltk.download('stopwords')
nltk.download('punkt')
```

```python
default_stopwords = set(nltk.corpus.stopwords.words('english'))
```

```python
string = 'Four score and seven years ago our fathers brought forth, upon this continent, a new nation, conceived in liberty, and dedicated to the proposition that all men are created equal. Now we are engaged in a great civil war, testing whether that nation, or any nation so conceived, and so dedicated, can long endure.'
words = nltk.word_tokenize(string)
words = [word.lower() for word in words if (len(word) > 1 and not word.isnumeric())]
words = [word for word in words if word not in default_stopwords]
words
```

```python
text = " ".join(words)
wordcloud = WordCloud().generate(text)
```

```python
import matplotlib.pyplot as plt
plt.imshow(wordcloud)
plt.show()
```
